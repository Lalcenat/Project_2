## Read Files

```python
# Read GDP File
gdp_dirty = pd.read_csv("gdp.csv")
gdp_dirty = gdp_dirty.set_index("Country Name")
```

```python
gdp_dirty = gdp_dirty.loc[:,'1996':'2016']
gdp_df = gdp_dirty.dropna(how = 'any')
gdp_df.head()
```

```python
life_dirty = life_dirty.loc[:,'1996':'2016']
life_df = life_dirty.dropna(how = 'any')
life_df.head()
```

```python
#Read Population File
pop_df = pd.read_csv("MergedPopDataCSV.csv")
pop_df = pop_df.iloc[:,:-2]
pop_df.rename(columns = {"Location":"Country Name"}, inplace = True)
pop_df = pop_df.drop_duplicates("Country Name")
pop_df = pop_df.set_index("Country Name")
pop_df
```

## Plot 1 - GDP growth across sub-regions (Dropdown menu)

```python
# Tranpose the DF
transposed_gdp = gdp_df.T
transposed_gdp.reset_index(inplace = True)
transposed_gdp.rename(columns = {"index":"year"}, inplace = True)
transposed_gdp
```

```python
#Getting the data per sub region - AP
AP_gdp_df = transposed_gdp[['year','East Asia & Pacific']]
AP_gdp_df.head()
```

```python
#Getting the data per sub region - LA
LA_gdp_df = transposed_gdp[['year','Latin America & Caribbean']]
LA_gdp_df.head()
```

```python
#Getting the data per sub region - NA
NA_gdp_df = transposed_gdp[['year','North America']]
NA_gdp_df.head()
```

```python
#Getting the data per sub region - MENA
MENA_gdp_df = transposed_gdp[['year','Middle East & North Africa']]
MENA_gdp_df.head()
```

```python
#Getting the data per sub region - SSA
SSA_gdp_df = transposed_gdp[['year','Sub-Saharan Africa']]
SSA_gdp_df.head()
```

```python
#plotting a line

ax = plt.gca()

#plotting the regions
transposed_gdp.plot(kind='line', x='year', y='European Union', ax=ax)
transposed_gdp.plot(kind='line', x='year', y='East Asia & Pacific', ax=ax)
transposed_gdp.plot(kind='line', x='year', y='Latin America & Caribbean', ax=ax)
transposed_gdp.plot(kind='line', x='year', y='North America', ax=ax)
transposed_gdp.plot(kind='line', x='year', y='Middle East & North Africa', ax=ax)
transposed_gdp.plot(kind='line', x='year', y='Sub-Saharan Africa', ax=ax)


#resizing image
plt.rcParams['figure.figsize'] = (16,9)

#setting labels
plt.title("GDP of World Regions")
plt.xlabel('Year')
plt.ylabel('GDP in $ Tens of Trillions')

#making legend fancy
plt.legend(fancybox=True, framealpha=1, shadow=True, borderpad=1)

plt.show()
```

We want to understand if population interferes in the amount of the GDP growth across sub-regions. In order to confirm our hypothesis, we will plot GDP Growth and GDP Growth per capta of each sub-region to graphically compare its growth.

```python
#Matching countries list
pop_ix = pop_df.index.tolist()
gdp_ix = gdp_df.index.tolist()

countries = set(gdp_ix).intersection(pop_ix)
countries
```

```python
#Create the GDP per capta df
gdp_fix = gdp_df.loc[countries]
gdp_fix = gdp_fix.sort_index()

pop_fix = pop_df.loc[countries]
pop_fix = pop_fix.sort_index()

gdp_percapta_df = gdp_fix.div(pop_fix)
gdp_percapta_df.head()
```

```python
#Transpose the DF
transposed_percapta = gdp_percapta_df.T
transposed_percapta.reset_index(inplace = True)
transposed_percapta.rename(columns = {"index":"year"}, inplace = True)
transposed_percapta
```

## Plot 5 - Scatter Plot GDP vs. Life Expectancy

```python
#Create a df of GDP per subregion
gdpsub_dfx = gdp_df.loc[["European Union", "East Asia & Pacific", "Latin America & Caribbean", "North America",\
                       "Middle East & North Africa", "Sub-Saharan Africa"]]
gdpsub_df = gdpsub_dfx.T
gdpsub_df
```

```python
#Create a df of Life Expectancy per subregion
lifesub_dfx = life_df.loc[["European Union", "East Asia & Pacific", "Latin America & Caribbean", "North America",\
                       "Middle East & North Africa", "Sub-Saharan Africa"]]
lifesub_df = lifesub_dfx.T
lifesub_df
```

```python
#Build the scatter plot
plt.scatter(gdpsub_df["European Union"], lifesub_df["European Union"], label = "European Union", color = "coral",\
           edgecolor = "black", alpha = 0.75, marker = "o")

plt.scatter(gdpsub_df["East Asia & Pacific"], lifesub_df["East Asia & Pacific"], label = "East Asia & Pacific",\
            color = "gold",edgecolor = "black", alpha = 0.75, marker = "o")

plt.scatter(gdpsub_df["Latin America & Caribbean"], lifesub_df["Latin America & Caribbean"], label = "Latin America & Caribbean",\
            color = "lightblue",edgecolor = "black", alpha = 0.75, marker = "o")

plt.scatter(gdpsub_df["North America"], lifesub_df["North America"], label = "North America",\
            color = "red",edgecolor = "black", alpha = 0.75, marker = "o")

plt.scatter(gdpsub_df["Middle East & North Africa"], lifesub_df["Middle East & North Africa"], label = "Middle East & North Africa",\
            color = "green",edgecolor = "black", alpha = 0.75, marker = "o")

plt.scatter(gdpsub_df["Sub-Saharan Africa"], lifesub_df["Sub-Saharan Africa"], label = "Sub-Saharan Africa",\
            color = "purple",edgecolor = "black", alpha = 0.75, marker = "o")


#Elements
plt.xlabel("GDP")
plt.ylabel("Life Expectancy")
plt.title("GDP vs Life Expectancy per Sub-region")

#Legend
plt.legend(fontsize = 8, title = "Sub-regions", loc = "best")

plt.show()
```
