```python
# Import the Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Read the gdp.csv

```python
gdp_df = pd.read_csv("gdp.csv")
gdp_df = gdp_df.set_index("Country Name")
gdp_df.head()
```

# Create a new data frame with the columns required

```python
gdp_df = gdp_df.loc[:,'1996':'2016']
gdp_df.head()
```

```python
# round the values
gdp_df = gdp_df.round()
gdp_df.head()
```

```python
# clean the cells that not have any value
gdp_df = gdp_df.dropna(how = 'any')
gdp_df.head()
```

# Read the life_expect.csv

```python
life_df = pd.read_csv('life_expect.csv')
life_df = life_df.set_index("Country Name")
life_df.head()
```

# Create a new data frame with the columns required

```python
life_df = life_df.loc[:,'1996':'2016']
life_df.head()
```

```python
# clean the cells that not have any value

life_df = life_df.dropna(how = 'any')
life_df.head()
```

# Generating a list of matching countries

```python
#saving a new DF to work with later
life1_df = life_df
gdp1_df = gdp_df
```

```python
gdp_ix = gdp_df.index.tolist()
life_ix = life_df.index.tolist()
```

```python
countries = set(gdp_ix).intersection(life_ix)
countries
```

## Read and manipulate the population file

```python
#read the file
dirty_pop = pd.read_csv("UNpopfile.csv")
dirty_pop.head()
```

```python
#check types
dirty_pop.dtypes
```

```python
#filter and remove unecessary columns
pop = dirty_pop.loc[dirty_pop["Variant"] == "Medium"]
pop = pop[["Location", "Time", "PopTotal"]]
pop = pop.set_index("Location")
pop.head()
```

```python
#match countries
pop_ix = pop.index.unique().tolist()
countries2 = set(pop_ix).intersection(countries)
countries2
```

```python
#match countries in population df
pop_df2 = pop.loc[countries2]
pop_df2 = pop_df2.sort_index()
pop_df2 = pop_df2.loc[pop_df2["Time"] > 1995]
pop_df2 = pop_df2.loc[pop_df2["Time"] < 2017]
pop_df = pop_df2.sort_values(["Location", "Time"])
pop_df = pop_df.reset_index()
pop_df.head()
```

```python
testx = pop_df.set_index("Time").T
testx
```

# Set the Index and rename Columns

```python
gdp_df = gdp_df.loc[countries]
gdp_df = gdp_df.sort_index()
gdp_df.columns = 'gdp_' + gdp_df.columns
gdp_df.reset_index(inplace=True)
gdp_df.rename(columns={"Country Name": "country_name"}, inplace=True)
gdp_df.head()
```

```python
life_df = life_df.loc[countries]
life_df = life_df.sort_index()
life_df.columns = 'lfe_' + life_df.columns
life_df.reset_index(inplace=True)
life_df.rename(columns={"Country Name": "country_name"}, inplace=True)
life_df.head()
```

# Percent Change

```python
#calculating percent change in gdp from 1996 and 2016
gdp_df["gdp_pct_change"] = (gdp_df["gdp_2016"]-gdp_df["gdp_1996"])/gdp_df["gdp_1996"]

gdp_df.head()
```

```python
#calculating percent change on life expectancy between 1996 and 2016
life_df["lfe_pct_change"] = (life_df["lfe_2016"]-life_df["lfe_1996"])/life_df["lfe_1996"]

life_df.head()
```

```python
#creating new dataframe on percent changes
life_gdp_pct = life_df[["country_name","lfe_pct_change"]]
life_gdp_pct["gdp_pct_change"] = gdp_df["gdp_pct_change"]

life_gdp_pct
```

```python
#transposing dataframe in order to plot a line graph

#drop pct change column from original gdp df
gdp_df.drop(columns=['gdp_pct_change'], inplace=True)

#gdp_df
#created new df where we set the index to country name in order to then transpose that with years
gdp_reindex = gdp_df.set_index('country_name')

#transposed the data frame
transposed_df = gdp_reindex.T

#reset the index to have the years as an actual column
transposed_df.reset_index(inplace=True)

#since we reset the index, we have to rename our new column that contains the years
transposed_df.rename(columns={"index": "year"}, inplace=True)


transposed_df
```

```python
#plotting a line

ax = plt.gca()

#plotting the regions
transposed_df.plot(kind='line', x='year', y='European Union', ax=ax)
transposed_df.plot(kind='line', x='year', y='East Asia & Pacific', ax=ax)
transposed_df.plot(kind='line', x='year', y='Latin America & Caribbean', ax=ax)
transposed_df.plot(kind='line', x='year', y='North America', ax=ax)
transposed_df.plot(kind='line', x='year', y='Middle East & North Africa', ax=ax)
transposed_df.plot(kind='line', x='year', y='Sub-Saharan Africa', ax=ax)


#resizing image
plt.rcParams['figure.figsize'] = (16,9)

#setting labels
plt.title("GDP of World Regions")
plt.xlabel('Year')
plt.ylabel('GDP in $ Tens of Trillions')

#making legend fancy
plt.legend(fancybox=True, framealpha=1, shadow=True, borderpad=1)

plt.show()
```

```python
#plotting scatterplot using % changes
ax = plt.scatter(life_gdp_pct["gdp_pct_change"], life_gdp_pct["lfe_pct_change"], marker="o", facecolors="blue", edgecolors="black", alpha = 0.50 )


fig_size = plt.rcParams["figure.figsize"]

fig_size[0] = 10
fig_size[1] = 10

plt.rcParams["figure.figsize"] = fig_size

plt.axis([-1,10,-0.05,0.5])
plt.grid()
plt.title("% change GDP vs % change life expectancy")
plt.xlabel("% change in GDP from 1996 to 2016")
plt.ylabel("% change in LE from 1996 to 2016")

z = np.polyfit(life_gdp_pct["gdp_pct_change"], life_gdp_pct["lfe_pct_change"], 1)
p = np.poly1d(z)
plt.plot(life_gdp_pct["gdp_pct_change"],p(life_gdp_pct["gdp_pct_change"]),"r--")

plt.savefig("../%change_metrics.png")

plt.show()
```
